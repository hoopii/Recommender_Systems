## Item-Based function

```python
# function with Pearson Correlation: 
def n_most_similar (movie_id ,n=10): 
  results_popularity  = rating_score.sort_values(by="total_score_m1", ascending=False).head(10)
  similar_to_id = user_movies_matrix.corrwith(user_movies_matrix[movie_id])
  corr_id = pd.DataFrame(similar_to_id, columns=['PearsonR'])
  corr_id.dropna(inplace=True)
  rating = pd.DataFrame(ratings.groupby('movieId')['rating'].mean())
  rating['rating_count'] = ratings.groupby('movieId')['rating'].count()
  id_corr_summary = corr_id.join(rating['rating_count'])
  id_corr_summary.drop(movie_id, inplace=True)
  top_n = id_corr_summary[id_corr_summary['rating_count']>=10].sort_values('PearsonR', ascending=False)
  top_n = top_n.merge(movies, left_index=True, right_on="movieId")
  # top_n = pd.merge(left=top_n, right=tags, how='left')
  top_n = top_n[['PearsonR', 'rating_count', 'movieId', 'title']]
  results_popularity  = rating_score.sort_values(by="total_score_m1", ascending=False).head(10)
  # get rid of the results from popularity ranking to not recommend movies twice: 
  top_n = top_n.loc[~top_n['movieId'].isin(results_popularity['movieId']), :]
  return top_n.head(n)
```

## User based approach: function

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def user_most_similar (user_id, n=5): 
  user_movies_ub = user_movies_matrix.copy()
  user_movies_ub.fillna(0, inplace=True)
  user_similarities = pd.DataFrame(cosine_similarity(user_movies_ub),
                                 columns=user_movies_ub.index, 
                                 index=user_movies_ub.index)
  weights = (
    user_similarities.query("userId!=@user_id")[user_id] / sum(user_similarities.query("userId!=@user_id")[user_id])
          )
  not_rated_movies = user_movies_ub.loc[user_movies_ub.index!=user_id, user_movies_ub.loc[user_id,:]==0]
  weighted_averages = pd.DataFrame(not_rated_movies.T.dot(weights), columns=["predicted_rating"])
  recommendations = weighted_averages.merge(movies, left_index=True, right_on="movieId")
  # get rid of results that have already been recommended in other systems
  results_popularity  = rating_score.sort_values(by="total_score_m1", ascending=False).head(10)
  recommendations = recommendations.loc[~recommendations['movieId'].isin(results_popularity['movieId']), :]
  recommendations = recommendations.loc[~recommendations['movieId'].isin(n_most_similar(356, 10)['movieId']), :]
  return  recommendations.sort_values("predicted_rating", ascending=False).head(n)
```

#### give a list of top-N movies recommended - alternative solution from documentation

```python
from collections import defaultdict
def get_top_n(predictions, n=10):
    """Return the top-N recommendation for each user from a set of predictions.

    Args:
        predictions(list of Prediction objects): The list of predictions, as
            returned by the test method of an algorithm.
        n(int): The number of recommendation to output for each user. Default
            is 10.

    Returns:
    A dict where keys are user (raw) ids and values are lists of tuples:
        [(raw item id, rating estimation), ...] of size n.
    """

    # First map the predictions to each user.
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    # Then sort the predictions for each user and retrieve the k highest ones.
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n
```

```python
def evaluation (data): 
  import numpy as np
  from sklearn.metrics import r2_score
  from sklearn.metrics import mean_absolute_error
  #  ratings positions in this dataset different than 0
  ratings_pos = pd.DataFrame(
      np.nonzero(np.array(data)), # find out all the positions different than 0
  ).T
  # split in train and test
  from sklearn.model_selection import train_test_split
  train_pos, test_pos = train_test_split(ratings_pos, 
                                       random_state=123, 
                                       test_size=.1)
  # create a train and a test with the information in those positions: 
  # 1: create an empty array full of 0, with the same shape as the users-items data
  train = np.zeros(data.shape)
  # 2: fill the set with the ratings based on the train positions
  for pos in train_pos.values: 
    index = pos[0]
    col = pos[1]
    train[index, col] = data.iloc[index, col]
  # 3: convert arrays to dataframe
  train = pd.DataFrame(train, 
                     columns=data.columns, 
                     index=data.index)
  # do the same with test data: 
  test = np.zeros(data.shape)
  for pos in test_pos.values: 
    index = pos[0]
    col = pos[1]
    test[index, col] = data.iloc[index, col]
  test = pd.DataFrame(test, 
                    columns=data.columns, 
                    index=data.index)
  # build similarity matrix for train set
  train_similarity = pd.DataFrame(cosine_similarity(train), 
                                columns=train.index, 
                                index=train.index)
  # define recommender function: 
  def recommender(index_pos, column_pos): 
    # build a df with the ratings for one place (column_name) and
    # the similarities to one user (index_name)
    results = (
      pd.DataFrame({
          'ratings': train.iloc[:,column_pos], 
          'similarities' : train_similarity.iloc[index_pos,:].tolist()
      })
    )
    # compute the weights
    results = results.assign(weights = results.similarities / (sum(results.similarities) -1))
    # compute the weighted ratings
    results = results.assign(weighted_ratings = results.ratings * results.weights)
    # return rating prediction for one user and one movie
    return results.weighted_ratings.sum()
 # use already defined recommender function to loop over all predictions for the test set and put them in a list 
  recs_test = []
  for row in test_pos.iterrows():
    recs_test.append(
        recommender(
            index_pos = int(row[1][0]), 
            column_pos = int(row[1][1])
        )
    )
  # make a df with positions of all the ratings in the test set,their true rating and prediction: 
  true_test_ratings = []
  for row in test_pos.iterrows():
    true_test_ratings.append(data.iloc[int(row[1][0]), int(row[1][1])])
  test_pos = test_pos.assign(true_rating = true_test_ratings, pred_rating = recs_test)

  # metrices: 
  mean_0_5 = test_pos.loc[test_pos.true_rating==0.5,:]["pred_rating"].mean()
  mean_1 = test_pos.loc[test_pos.true_rating==1,:]["pred_rating"].mean()
  mean_1_5 = test_pos.loc[test_pos.true_rating==1.5,:]["pred_rating"].mean()
  mean_2 = test_pos.loc[test_pos.true_rating==2,:]["pred_rating"].mean()
  mean_2_5 = test_pos.loc[test_pos.true_rating==2.5,:]["pred_rating"].mean()
  mean_3 = test_pos.loc[test_pos.true_rating==3,:]["pred_rating"].mean()
  mean_3_5 = test_pos.loc[test_pos.true_rating==3.5,:]["pred_rating"].mean()
  mean_4 = test_pos.loc[test_pos.true_rating==4,:]["pred_rating"].mean()
  mean_4_5 = test_pos.loc[test_pos.true_rating==4.5,:]["pred_rating"].mean()
  mean_5 = test_pos.loc[test_pos.true_rating==5,:]["pred_rating"].mean()
  mae = mean_absolute_error(test_pos.true_rating, test_pos.pred_rating)
  r2 = r2_score(test_pos.true_rating, test_pos.pred_rating)
  return test_pos, print(f'''mean for 0.5: {mean_0_5}
  mean for 1: {mean_1} 
  mean for 1,5: {mean_1_5} 
  mean for 2: {mean_2} 
  mean for 2.5: {mean_2_5}
  mean for 3: {mean_3} 
  mean for 3.5: {mean_3_5} 
  mean for 4: {mean_4} 
  mean for 4.5: {mean_4_5} 
  mean for 5: {mean_5} 
  mean absolute error {mae} 
  r2_score:{r2}''')
```
